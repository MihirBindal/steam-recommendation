```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

```python
steam = pd.read_csv("steam.csv")
steamspy_tag_data = pd.read_csv("steamspy_tag_data.csv")
```

```python
small = ['0-20000','20000-50000','50000-100000']
huge = ['5000000-10000000','100000000-200000000','20000000-50000000','50000000-100000000']
rows = steam.shape[0]
for i in range(0,rows):
    if steam['owners'][i] in small:
        steam['owners'][i]='small'
    elif steam['owners'][i] in huge:
        steam['owners'][i]='huge'
    else:
        steam['owners'][i]='medium'
```

```python
steam.head()
```

```python
y = pd.get_dummies(steam.owners, prefix='size')
y=y.drop('size_medium', axis=1)
y.head()
```

```python
for i in range(0,rows):
    if steam['required_age'][i] >= 18:
        steam['required_age'][i]=1
    else:
        steam['required_age'][i]=0
```

```python
z = pd.get_dummies(steam.required_age, prefix='age_verification')
z=z.drop('age_verification_0', axis=1)
z.rename(columns = {'age_verification_1':'age_verification'}, inplace = True)
z.head()
```

```python
to_drop =['name','owners','release_date','developer', 'publisher', 'platforms', 'categories', 'genres', 'steamspy_tags',
          'achievements', 'english', 'required_age']
steam=steam.drop(to_drop, axis=1)
steam.head()
```

```python
steam.shape
```

```python
steam = pd.concat([steam, z], axis=1, sort=False)
```

```python
steam = pd.concat([steam, y], axis=1, sort=False)
```

```python
steam.head()
```

```python
steam.shape
```

```python
final_df=pd.merge(left=steam, right=steamspy_tag_data, how='left', left_on='appid', right_on='appid')
```

```python
final_df.head()
```

```python
final_df=final_df.drop('appid', axis=1)
```

```python
final_df.shape
```

```python
scaler=MinMaxScaler(feature_range=(0, 1))
features=scaler.fit_transform(final_df)
```

```python
features
```

```python
import numpy as np
np.round(features,3)
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
model = NearestNeighbors(n_neighbors=6).fit(features)
```

```python
distances, indices = model.kneighbors(features)
```

```python
steam_original = pd.read_csv("steam.csv")
def get_index_from_name(name):
    return steam_original[steam_original["name"]==name].index.tolist()[0]
```

```python
all_game_names = list(steam_original.name.values)
```

```python
def print_similar_games(query):
        found_id = get_index_from_name(query)
        for id in indices[found_id][1:]:
            print(steam_original.iloc[id]["name"])
```

```python
print_similar_games("Bio Inc. Redemption")
```
